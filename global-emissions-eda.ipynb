**3. Add Filters to Select Specific data for EDA**

```python
region = df[df['region']!='World']
region
```

```python
region = region.groupby(region['region']).sum('emissions')
r = region.reset_index()
```

```python
text = r['emissions'].apply(lambda x: '{:.2f}'.format(x))
```

```python
fig = px.bar(r, x='region', y='emissions',text=text,color='region', width=1000, height=800,title='Regions vs Total Emissions')
fig.update_traces(textposition='outside',textfont=dict(size=18))
fig.show()
```

**So, We can see the Asia Pacific region emits highest emission in compare to other region. For this chat I used plotly library and use bar chart with text**

```python
emissions = df[df['region']!='World']
emissions
```

```python
past_emissions = emissions.groupby(['country','baseYear']).sum('emissions')
ps = past_emissions.reset_index()
```

```python
fig = px.choropleth(ps, locations='country', locationmode='country names',
                    color='emissions', hover_name='country',
                    projection='miller',width=800, height=800 )
fig.show()
```

**Upper map chart is showing total emissions by country. Here, we can see China has the highest methane emissions compared to other countries, and that the USA and Russia are close behind. India, Iran, and Indonesia are also significant emitters of methane.**

```python
df
```

**Here, We select emission type is Agriculture and we exclude region world.**

```python
a_type= df[(df['type'] == 'Agriculture') & (df['region'] != 'World')]
a_type
```

```python
Agriculture = a_type.groupby(['region','country','emissions'])
Agriculture=Agriculture.first()
```

```python
Agriculture = Agriculture.reset_index()
Agriculture
```

```python
africa_data = Agriculture[Agriculture['region'] == 'Africa']
africa_data
```

```python
fig = px.choropleth(africa_data, locations="country",locationmode='country names', color='emissions',
                    color_continuous_scale='oryel', range_color=(0, africa_data['emissions'].max()),
                    hover_name='country', projection='mercator',title='Total Emissions by African Countries ')

fig.update_geos(center=dict(lon=15, lat=0), 
                projection_rotation=dict(lon=0, lat=0, roll=0),
                scope="africa", 
                lataxis_range=[-40, 40], 
                lonaxis_range=[-40, 70])

fig.show()
```

**Here, I perform choropleth map visualisation on specific regions like Africa, Asia, Europe, North America and South America by using plotly library. Here I done only for emission type is Agriculture. you can Perform for other types such as Energy and West**

```python
Agriculture['region'] = Agriculture['region'].replace('Middle East', 'Asia')
Agriculture['region'] = Agriculture['region'].replace('Asia Pacific', 'Asia')
Agriculture['region'] = Agriculture['region'].replace('Russia & Caspian', 'Asia')
```

```python
asia_data = Agriculture[Agriculture['region'] == 'Asia']
asia_data
```

```python
fig = px.choropleth(asia_data, locations="country",locationmode='country names', color='emissions',
                    color_continuous_scale='curl', range_color=(0, asia_data['emissions'].max()),
                    hover_name='country', projection='mercator',title='Total Emissions by Asian Countries')

fig.update_geos(center=dict(lon=100, lat=30), 
                projection_rotation=dict(lon=0, lat=0, roll=0),
                scope="asia", 
                lataxis_range=[0, 60], 
                lonaxis_range=[60, 180])


fig.show()
```

```python
Europe_data = Agriculture[Agriculture['region'] == 'Europe']
Europe_data = Europe_data.drop(Europe_data[(Europe_data['country'] == 'European Union') | (Europe_data['country'] == 'Other EU17 countries') | (Europe_data['country'] == 'Other EU7 countries') | (Europe_data['country'] == 'Other countries in Europe')].index)
Europe_data
```

```python
fig = px.choropleth(Europe_data, locations="country",locationmode='country names', color='emissions',
                    color_continuous_scale='magenta', range_color=(0, Europe_data['emissions'].max()),
                    hover_name='country', projection='mercator',title='Total Emissions by European Countries')

fig.update_geos(center=dict(lon=15, lat=50),
                projection_rotation=dict(lon=0, lat=0, roll=0),
                scope="europe",
                lataxis_range=[30, 60],
                lonaxis_range=[-30, 50])


fig.show()
```

```python
north_america_data = Agriculture[Agriculture['region'] == 'North America']
north_america_data
```

```python
fig = px.choropleth(north_america_data, locations="country",locationmode='country names', color='emissions',
                    color_continuous_scale='sunsetdark', range_color=(0, north_america_data['emissions'].max()),
                    hover_name='country', projection='mercator',title='Total Emissions by North American Countries')


fig.update_geos(center=dict(lon=-100, lat=45),
                projection_rotation=dict(lon=-100, lat=0, roll=0),
                scope="north america",
                lataxis_range=[0, 80],
                lonaxis_range=[-180, -20])


fig.show()
```

```python
south_america_data = Agriculture[Agriculture['region'] == 'Central and South America']
south_america_data
```

```python
df1 = df[df['region']=='World']
```

```python
df1['segment'].unique()
```

```python
df3 = df1.groupby(['segment']).sum('emissions')
df3 = df3.reset_index()
df3 = df3.drop(df3[(df3['segment'] == 'Total')].index)
df3
```

```python
fig = px.pie(df3, values='emissions', names='segment',hole=.5)
fig.update_layout(
    title='Distribution of Energy West',
    annotations=[dict(text='Energy West',
                      font=dict(size=20),
                      showarrow=False,
                      x=0.5,
                      y=0.5)]
)
fig.show()
```
